```python
import pandas as pd
from pathlib import Path

def check_daily_files(base_dir: str):
    base = Path(base_dir)
    for f in base.rglob("date=*.parquet"):
        try:
            df = pd.read_parquet(f)
            n = len(df)
            status = "OK" if n == 1440 else f"MISSING ({n})"
        except Exception as e:
            status = f"ERROR ({e})"
        print(f"{f} -> {status}")

# 範例：換成你本機的路徑
check_daily_files(r"C:\Users\Burkey\LetsGetStarted\data")
```

```python
import requests, pandas as pd

def to_ms(ts): return int(pd.Timestamp(ts, tz="UTC").timestamp()*1000)

sym = "BTCUSDT"
t0 = "2023-03-24 00:00"
t1 = "2023-03-25 00:00"

def fetch(start, end):
    r = requests.get("https://api.binance.com/api/v3/klines", params={
        "symbol": sym, "interval":"1m", "startTime": to_ms(start), "endTime": to_ms(end)-1, "limit":1000
    }, timeout=30)
    r.raise_for_status()
    return r.json()

# 前 1000 分鐘（~16h40m）
mid = pd.Timestamp(t0, tz="UTC") + pd.Timedelta(minutes=1000)
p1 = fetch(t0, mid.isoformat())
# 後 440 分鐘
p2 = fetch(mid.isoformat(), t1)
print("api_rows =", len(p1)+len(p2))
```
